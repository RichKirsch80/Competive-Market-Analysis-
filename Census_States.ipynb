```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading in File from Census.gov CSV keeping only relevant rows out of the 60 available
pred = lambda x: x not in [0, 1, 27, 32, 45]
census_data_df = pd.read_csv("QuickFacts_Aug-06-2021.csv", skiprows=pred, index_col=0)
census_data_df
```

```python
# dropping unnecessary columns
census_data_df = census_data_df.drop(["Fact Note", "Value Note for Lake Forest city, Illinois",
                                     "Value Note for Mundelein village, Illinois", "Value Note for Lincolnshire village, Illinois",
                                     "Value Note for Libertyville village, Illinois", "Value Note for Vernon Hills village, Illinois",
                                     "Value Note for Lake County, Illinois"], axis=1)
```
